# GenAI Assignment – Prompt Templates using LangChain

## Objective:
To build dynamic and reusable prompt systems using LangChain instead of hardcoded prompts.

## Introduction:
This assignment focuses on creating structured and reusable prompt templates using LangChain.
Instead of writing static prompts, we build a system that dynamically generates prompts based on user inputs.

```python
# Install LangChain (latest version)
!pip install -U langchain langchain-core

# Import required classes
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate
```

## Task 1: Replace Hardcoded Prompts

📌 Explanation:

We replace static f-string prompts with reusable PromptTemplate.

```python
# Creating a reusable prompt template
template = "Explain {topic} in simple terms for beginners"

# Initialize PromptTemplate
prompt = PromptTemplate(
    input_variables=["topic"],
    template=template
)

# Generate output
print(prompt.format(topic="Artificial Intelligence"))
```

#TASK 2: Multi-Input Prompt System
📌 Explanation:

We use multiple input variables like topic, audience, and tone.

```python
# Multi-input template
template = "Explain {topic} for {audience} in a {tone} tone"

# Create PromptTemplate
prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# Test cases
print(prompt.format(topic="AI", audience="beginners", tone="friendly"))
print(prompt.format(topic="Python", audience="kids", tone="fun"))
print(prompt.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

#TASK 3: Prompt Variations Engine
📌 Explanation:

We create multiple templates for different styles.

```python
# Teaching style template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview style template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Storytelling template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Test with same topic
topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

#TASK 4: ChatPromptTemplate System
📌 Explanation:

We simulate chat-based prompts using roles.

```python
# Create chat prompt template
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}"),
    ("human", "Explain {topic}")
])

# Example usage
print(chat_template.format(role="teacher", topic="Neural Networks"))
print(chat_template.format(role="interviewer", topic="AI"))
print(chat_template.format(role="motivator", topic="Success"))
```

#TASK 5: Input Validation Layer
📌 Explanation:

We validate user inputs and assign defaults if invalid.

```python
# Function to validate inputs
def validate_inputs(audience, tone):

    # Allowed values
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Check audience
    if audience not in valid_audience:
        print("Invalid audience! Setting default = beginner")
        audience = "beginner"

    # Check tone
    if tone not in valid_tone:
        print("Invalid tone! Setting default = formal")
        tone = "formal"

    return audience, tone

# Test validation
print(validate_inputs("alien", "funny"))
```

#TASK 6: Prompt Generator App
📌 Explanation:

We combine everything into one function.

```python
def generate_prompt(topic, audience, tone, style):

    # Step 1: Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Select template based on style
    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} teaching style"

    elif style == "interview":
        template = "Ask interview questions about {topic} for {audience} in a {tone} tone"

    elif style == "storytelling":
        template = "Explain {topic} for {audience} in a {tone} storytelling style"

    else:
        template = "Explain {topic} for {audience} in a {tone} tone"

    # Step 3: Create PromptTemplate
    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    # Step 4: Generate output
    return prompt.format(topic=topic, audience=audience, tone=tone)


# Test the function
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

#TASK 7: Template Reusability Test
📌 Explanation:

We use one template with multiple inputs.

```python
# Single reusable template
template = "Explain {topic} for {audience} in a {tone} tone"

prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# Test with different inputs
print(prompt.format(topic="AI", audience="beginner", tone="fun"))
print(prompt.format(topic="Python", audience="kids", tone="casual"))
print(prompt.format(topic="Machine Learning", audience="expert", tone="formal"))
print(prompt.format(topic="Deep Learning", audience="student", tone="fun"))
print(prompt.format(topic="NLP", audience="engineer", tone="formal"))
```

#FINAL PIPELINE
📌 Explanation:

Complete flow from input → validation → prompt generation

```python
# Final pipeline execution

topic = "Neural Networks"
audience = "beginner"
tone = "fun"
style = "storytelling"

result = generate_prompt(topic, audience, tone, style)

print("Final Generated Prompt:")
print(result)
```
